##**Demandas de ML**

###**¿Cuales son los valores promedio de alquiler por tipo de por propiedad?**

```python
datos.head()
```

```python
datos.groupby("Tipo")["Valor"].mean()
```

```python
datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
```

```python
df_tipo_precio = datos.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")

df_tipo_precio.plot.bar(title="Precio promedio por tipo de propiedad", color="green")
```

```python
df_tipo_precio.plot(kind="barh", figsize=(12,8), color="green")
```

```python
datos.Tipo.unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala', 'Edificio Completo','Tienda/Salón', 'Casa Comercial', 'Terreno Estándar', 'Cochera/Estacionamiento', 'Galpón/Depósito/Almacén', 'Tienda en Centro Comercial',
       'Hotel', 'Loteo/Condominio','Industria']
```

```python
datos.query("@inmuebles_comerciales in Tipo")
```

```python
datos.query("@inmuebles_comerciales not in Tipo")
```

```python
df = datos.query("@inmuebles_comerciales not in Tipo")
df.head()
```

```python
df_tipo_precio = df.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_tipo_precio.plot(kind="barh", figsize=(12,8), color="green")
```
